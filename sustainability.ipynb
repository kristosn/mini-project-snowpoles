```python
import numpy as np
```

```python
a100_power_consumption = 250 # Watt
# https://bil24.no/tesla-model-y-long-range-firehjulsdrift-nordmenns-helt-klare-favoritt
tesla_kwh_per_km = 0.16
print(tesla_kwh_per_km)
```

Training times

```python
def seconds_to_hours(seconds):
    return seconds / 3600

yolo_training_time_v1 = 769.24 # seconds
yolo_traning_time_iphone = 1500 # seconds
rfdetr_training_time_v1 = 0
rfdetr_training_time_iphone = 0

yolo_predict_time_v1 = 5.24
yolo_predict_time_iphone = 9.05

total_training_time = yolo_training_time_v1 + yolo_traning_time_iphone + rfdetr_training_time_v1 + rfdetr_training_time_iphone + yolo_predict_time_v1 + yolo_predict_time_iphone# s

total_training_time_hours = seconds_to_hours(total_training_time)

print(total_training_time_hours)
```

Energy calculations

```python
def joule_to_kwh(joule):
    return joule / 3_600_000

total_energy = joule_to_kwh(a100_power_consumption * total_training_time) # kwh

print(total_energy) 


```

km calculations

```python
km_tesla = total_energy / tesla_kwh_per_km # km

print(f"You can drive {km_tesla:.2f} km with the amount of energy used training the models on both datasets with the A100 NVIDIA GPU." )
```
