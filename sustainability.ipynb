```python
import numpy as np
```

```python
a100_power_consumption = 250 # Watt
# https://bil24.no/tesla-model-y-long-range-firehjulsdrift-nordmenns-helt-klare-favoritt
tesla_kwh_per_km = np.mean([15.1, 20.1, 22.8]) / 100
print(tesla_kwh_per_km)
```

Training times

```python


yolo_training_time_v1 = 769.24 # seconds
yolo_traning_time_iphone = 1500 # seconds
rfdetr_training_time_v1 = None
rfdetr_training_time_iphone = None




```

Energy calculations

```python
def joule_to_kwh(joule):
    return joule / 3_600_000

energy_yolo_v1 = joule_to_kwh(a100_power_consumption * yolo_training_time_v1)
energy_yolo_iphone = joule_to_kwh(a100_power_consumption * yolo_traning_time_iphone)

print(energy_yolo_v1, energy_yolo_iphone)


```

km calculations

```python
km_tesla_yolo_v1 = energy_yolo_v1 / tesla_kwh_per_km
km_tesla_yolo_iphone = energy_yolo_iphone / tesla_kwh_per_km

print(km_tesla_yolo_v1, km_tesla_yolo_iphone)
```
