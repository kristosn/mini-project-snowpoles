# Sustainability
Estimating energy consumption from model training and inference, compared to Tesla model Y driving distance.

```python
# Hardware & vehicle constants
A100_POWER_W     = 250   # Watt
TESLA_KWH_PER_KM = 0.16  # kWh/km
```

## Training & Prediction Times

```python
# All times in seconds
training_times = {
    "yolo_v1":        769.24,
    "yolo_iphone":   1500.00,
    "rfdetr_v1":     3143.49,
    "rfdetr_iphone": 13952.08,
}

prediction_times = {
    "yolo_v1":        5.24,
    "yolo_iphone":    9.05,
    "rfdetr_v1":      6.15,
    "rfdetr_iphone": 12.57,
}

total_seconds = sum(training_times.values()) + sum(prediction_times.values())
print(f"Total compute time: {total_seconds / 3600:.2f} hours")
```

## Energy Consumption

```python
total_energy_kwh = (A100_POWER_W * total_seconds) / 3_600_000  # W*s -> kWh
print(f"Total energy used: {total_energy_kwh:.2f} kWh")
```

## Equivalent Driving Distance

```python
km_equivalent = total_energy_kwh / TESLA_KWH_PER_KM
print(
    f"The energy used is equivalent to driving {km_equivalent:.2f} km "
    f"in a Tesla (A100 GPU)."
)
```
